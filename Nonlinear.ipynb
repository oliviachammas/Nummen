c) Skriv ett Python-program som beräknar de nollställen som fixpunktsiterationen ovan kan hitta till (1). Programmet ska skriva ut xn och skillnaden |xn+1 − xn| efter varje iteration, samt returnera ett svar med ett fel som är mindre än en given tolerans τ = 10E−10. Använd
lämpligt avbrottsvillkor för att säkerställa detta.

```python
#Fixpunktsmetoden
def gfun(x): #definierar funktionen g
    g = (1/20)*x**2 + (11/20)*x -3/5*np.sin(3*x +1) + 1
    return g

def fixp(x):
    tol = 1E-10 #toleransen given
    diff = 1000 #sätter den som ett värde
    it = 0 #antal interationer, startar från noll
    while diff>tol and it<1000: #denna while loop kommer fortsätta så länge diff>tol och antal it är under 1000. 
        it += 1 #antal iterationer ökar med en varje gång
        xold = x #gamla x-värdet blir nya
        x = gfun(xold) #fixpunktsmetoden
        diff = abs(x - xold) #diffen, denna ska minska för att fixpunktsmetoden ska stämma
        print(it, x, diff)
    return it, x, diff

x = [2, 4, 6.0] #då vi vet att dessa konvergerar enligt fixpunktsmetoden kan vi endast använda dessa
for i in x: #för varje värde i x ansätter vi då denna fixpunktsmetod
    print(fixp(i))
    print('') #separator
```

e) Utöka programmet ovan så att det beräknar samtliga nollställen till (1) med hjälp av
Newtons metod. Programmet ska skriva ut xn och skillnaden |xn+1 −xn| efter varje
iteration, samt returnera ett svar med ett fel som är mindre än en given tolerans τ =
10E−10. Använd lämpligt avbrottsvillkor för att säkerställa detta.

```python
#Newtons metod
def func(x):
    f = x**2 - 9*x - 12*np.sin(3*x + 1) + 20
    fp = 2*x - 9 -36*np.cos(3*x + 1)
    return f, fp

def newton(x):
    tol = 1E-10 # given tolerans
    diff = 1 # initiera
    it = 0 # starta räknare
    while diff > tol and it < 1000:
        it += 1
        f, fp = func(x)
        xnew = x - f/fp
        diff = abs(xnew-x)
        x = xnew # uppdatera 
        print(it, x, diff)
    return it, x, diff

    
x = [1.95, 2.72, 3.86, 4.90, 6.0, 6.83]
for i in x:
    print(newton(i))
    print('') #separator, lättare att urskilja
```
