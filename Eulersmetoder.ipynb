```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Ex från föreläsning
def Fram_Euler(y):
    f = np.sin(3*t) - 2*y 
    return f

t0 = 0 # Begynnelsetid
T = 8 # Sluttid 
n = 50 # Antal steg (delintervall) 
h = T/n # Steglängd 
y0 = 1.2 # Begynnelsevärde
tVec = np.linspace(t0,T,n+1) # t-värden för vilka lösning beräknas
yVec = np.empty(tVec.size) # spara lösningen för varje tidpunkt
y = y0 # Initiera y till begynnelsevärdet 
yVec[0] = y # Spara y0 på första platsen i yVec
i = 0 

for t in tVec[:-1]: # Lösningen i n ges av y och t i n-1
    y = y + h*Fram_Euler(y) # Steg från n till n+1 med Euler 
    i += 1 
    yVec[i] = y # Spara lösningen
    
t = np.linspace(0, 8, 100)
y = (93/65)*np.e**(-2*t) - (3/13)*np.cos(3*t) + (2/13)*np.sin(3*t)
```

```python
plt.plot(tVec,yVec,'o')
plt.plot(t,y)
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0, 8, -0.75, 1.3])
plt.show()
```
